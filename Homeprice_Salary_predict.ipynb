**Handling NA values**

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
import math
df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
```

```python
df
```

```python
model = linear_model.LinearRegression()
model.fit(df.drop('price',axis= 'columns'),df.price)
```

```python
model.predict([[3000,3,1]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
df2 = pd.read_csv('hiring.csv')
df2
```

```python
df2.experience = df2.experience.fillna("zero")
```

```python
df2
```

```python
from word2number import w2n
```

```python
df2.experience = df2.experience.apply(w2n.word_to_num)
df2
```

```python
median_test_scor = math.floor(df2['test_score(out of 10)'].mean())
```

```python
median_test_scor
```

```python
df2['test_score(out of 10)'] = df2['test_score(out of 10)'].fillna(median_test_scor)
```

```python
df2
```

```python
model2 = linear_model.LinearRegression()
model2.fit(df2[['experience','test_score(out of 10)','interview_score(out of 10)']],df2['salary($)'])
```

```python
model2.predict([[2,9,6]])
```
